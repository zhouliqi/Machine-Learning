# 探索tips数据集

```python
# 导入我们所需要的包
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

tips数据集是一个餐厅侍者收集的关于小费的数据，其中包含了七个特征；总费用，
付小费的金额，付款人的性别，是否吸烟，日期，日间，顾客人数。

```python
tips = sns.load_dataset("tips")
print(len(tips))
tips.head()
```

```python
sns.distplot(tips["total_bill"]); # kde=False
```

由上图可知顾客小费的总金额主要在5-35范围之间

```python
# count plot图
sns.countplot(x="smoker", data=tips);
```

来餐厅就餐的顾客中，吸烟者的人数较多一点

```python
sns.countplot(x="time", data=tips);
```

上图表明来餐厅吃晚饭的人多一些，来吃午餐的人少一些

```python
sns.countplot(x="day", data=tips);
```

星期五来餐厅的人会少一些

```python
sns.countplot(x="size", data=tips);
```

两个人来餐厅就餐的总次数更多

```python
# joint plot图
sns.jointplot(x="total_bill", y="tip", data=tips);
```

上图可以看出，顾客消费水平在10-30元之间，对应给侍者的小费为1-5元之间

```python
sns.jointplot(x="total_bill", y="tip", kind="hex", data=tips);
```

```python
sns.jointplot(x="total_bill", y="tip", kind="reg", data=tips);
```

通过做一条简单的回归线，它表明了小费的金额是随着总账单金额的增加而增加的。

```python
# 某个区域越暗，表明这个区域对应的频次越多。
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde");
```

```python
# pair plot图
sns.pairplot(tips);
```

```python
sns.pairplot(tips, hue="sex", markers=["o", "s"]);
```

### rug plot图

```python
# 消费金额在各个值上的边缘分布
sns.rugplot(tips["total_bill"]);
```

### kde plot 图

```python
# 显示各个消费总金额数值的统计分布
sns.kdeplot(tips['total_bill'], shade=True);
```

```python
sns.barplot(x="sex", y="total_bill",  data=tips);
```

男性买单次数会多一些

```python
sns.boxplot(x="day", y="total_bill", data=tips);
```

上图表示大部分账单是在周六和周日支付的

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex");
```

上图表示周六女性买单次数比男性多

### violin plot

```python
sns.violinplot(x="day", y="total_bill", data=tips);
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,
               hue="sex", split=True);
```

### strip plot图

```python
sns.stripplot(x="day", y="total_bill", jitter=False, data=tips);
```

```python
sns.stripplot(x="day", y="total_bill", 
              data=tips, jitter=True,
              hue="sex", dodge=True);
```

### swarm plot图

```python
sns.swarmplot(x='day', y='total_bill', data=tips);
```

swarm plot不会重叠数据点

### heat map图

```python
sns.heatmap(tips.corr());
```

显示特征的相关性，在这里两个特征的颜色越浅，相关性越强

### pair grid图

```python
g = sns.PairGrid(tips)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot);
```
