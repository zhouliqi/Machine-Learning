```python
# 4.4.2　用plt.scatter画散点图
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.randn(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar(); # 显示颜色条
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel('iris.features_name[0]')
plt.ylabel('iris.features_name[1]');
```

```python
# 4.4.3 plot与scatter：效率对比
# 由于 plt.scatter 会对每个散点进行单独的大小与颜色的渲染，因此渲染器会消耗更多的资源。
# 而在 plt.plot 中，散点基本都彼此复制
# 因此面对大型数据集时，plt.plot 方法比 plt.scatter 方法好。
```

```python
# 4.5.1　基本误差线
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

```python
from sklearn.gaussian_process import GaussianProcess

model = lambda x: x * np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

# 定义高斯过程拟合结果
gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4,thetaU=1E-1,
                     random_start=100)
gp.fit(xdata[:, np.newaxis], ydata)
xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE) # 获得2*sigma 95%置信区间

# 结果可视化
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                 color='gray', alpha=0.2)
plt.xlim(0, 10);
```

```python
# 4.6　密度图与等高线图
# 三维函数的可视化
def f(x, y):
    return np.sin(x)**10 + np.cos(10 + y * x) * np.cos(x)

# 等高线图可以用 plt.contour 函数来创建。
# 它需要三个参数：x 轴、y 轴、z 轴三个坐标轴的网格数据。
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X,Y, Z, colors='black');
```

```python
def fx(x, y):
    return np.sqrt(np.power(x, 2) * np.power(y, 2))
x = np.linspace(0,100,500)
y = np.linspace(0,100,400)
X, Y = np.meshgrid(x, y)
Z = fx(X, Y)
plt.contour(X, Y, Z, colors='black');
```
